```python
def bubble_sort(array):
    counter = 0
    sorted = False
    while not sorted:
        sorted = True
        for i in range(len(array)-1):
            counter += 1
            if array[i] > array[i+1]:
                array[i], array[i+1] = array[i+1], array[i]
                if sorted:
                    sorted = False
    return counter
```

```python
def insertion_sort(array):
    counter = 0
    aux = 0
    j = 0
    for i in range(1, len(array)):
        aux = array[i]
        j = i-1
        while j >= 0 and aux < array[j]:
            counter += 1
            array[j+1] = array[j]
            j -= 1
        array[j+1] = aux
    return counter
```

```python
def merge(array1, array2):
    counter = 0
    aux_array = [0]*(len(array1)+len(array2))
    index1, index2, index_aux = 0, 0, 0

    while index1 < len(array1) and index2 < len(array2):
        counter += 1
        if array1[index1] < array2[index2]:
            aux_array[index_aux] = array1[index1]
            index1 += 1
        else:
            aux_array[index_aux] = array2[index2]
            index2 += 1
        index_aux += 1

    while index1 < len(array1):
        counter += 1
        aux_array[index_aux] = array1[index1]
        index1 += 1
        index_aux += 1

    while index2 < len(array2):
        counter += 1
        aux_array[index_aux] = array2[index2]
        index2 += 1
        index_aux += 1
    return aux_array, counter

def mergesort(array):
    counter, aux_counter = 1, 0
    if len(array) > 1:
        half = len(array)//2
        left_array, right_array = array[:half], array[half:]
        
        left_array, aux_counter = mergesort(left_array)
        counter += aux_counter

        right_array, aux_counter = mergesort(right_array)
        counter += aux_counter

        array, aux_counter = merge(left_array, right_array)
        counter += aux_counter

    return array, counter
    
def merge_sort(array):
    _, a = mergesort(array)
    return a
```

```python
def partition(array, start, end):
    pivot = array[end]
    lower = start
    upper = end-1
    finished = False

    counter = 0

    while not finished:
        while not finished:
            counter += 1
            if lower == upper:
                finished = True
                break
            if array[lower] > pivot:
                array[upper] = array[lower]
                break
            lower += 1
        while not finished:
            counter += 1
            if upper == lower:
                finished = True
                break
            if array[upper] < pivot:
                array[lower] = array[upper]
                break
            upper -= 1
    array[upper] = pivot
    return upper, counter

def quicksort(array, start, end):
    if start >= end:
        return 0

    counter = 1
    pivot, aux_counter = partition(array, start, end)
    counter += aux_counter

    aux_counter = quicksort(array, start, pivot-1)
    counter += aux_counter
    aux_counter = quicksort(array, pivot+1, end)
    counter += aux_counter
    
    return counter
    
def quick_sort(array):
    return quicksort(array, 0, len(array)-1)
```

```python
def heapify(heap, n, root):
    largest = root
    left = 2*root+1
    right = 2*root+2
    counter = 1
    if left < n and heap[left] > heap[largest]:
        largest = left
        counter += 1
    if right < n and heap[right] > heap[largest]:
        largest = right
    counter += 1
    if largest != root:
        heap[root], heap[largest] = heap[largest], heap[root]
        heapify(heap, n, largest)
    return counter 

def heap_sort(array):
    counter, aux_counter = 0, 0
    for i in range(len(array)//2-1, -1, -1):
        aux_counter = heapify(array, len(array), i)
        counter += aux_counter
    for i in range(len(array)-1, 0, -1):
        array[i], array[0] = array[0], array[i]
        aux_counter = heapify(array, i, 0)
        counter += aux_counter
    return counter
```
